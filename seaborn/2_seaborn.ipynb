```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('C:\\Users\\admin\\Desktop\\data_science\\data\\bigmart_data.csv')
df=pd.DataFrame(data)

df=df.dropna(how='any')
df['Visibility_Scaled']=df['Item_Visibility']*100
```
