```python
df_mart=pd.read_csv('C:\\Users\\admin\\Desktop\\Data Science\\Data for ananlysis\\bigmart_data.csv')
df_mart.head(3)
```

```python
df_mart.info()
```

```python
df_mart.dropna(how='any',inplace=True)
df_mart.info()
```

```python
df_mart.head(3)
```

```python
sales=df_mart.groupby('Outlet_Establishment_Year').Item_Outlet_Sales.mean()
sales
```

```python
items=df_mart.groupby('Item_Type').Item_MRP.mean()
items
```

```python
x=items.index.to_list()
y=items.to_list()
```

```python
plt.plot(x,y,marker='*',color='r', linestyle='--')
#plt.figure(figsize=(10,12))
plt.title('BiG Basket')
plt.xlabel('Item Type')
plt.ylabel('Item Price')
plt.xticks(labels=x, ticks=np.arange(len(x)))
```

```python
plt.bar(x,y, color=['r','y','g'])
#plt.figure(figsize=(10,12))
plt.title('BiG Basket')
plt.xlabel('Item Type')
plt.ylabel('Item Price')
plt.xticks(labels=x, ticks=np.arange(len(x)))
```
