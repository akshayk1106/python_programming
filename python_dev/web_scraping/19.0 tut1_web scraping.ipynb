```python
wiki=requests.get('https://en.wikipedia.org/wiki/World_War_II')
soup=BeautifulSoup(wiki.text,'html')
print(soup.find('title').text)
```

```python
print(soup.prettify())
```

```python
contents=soup.find_all('div',class_='toc')
for i in contents:
    print(i.text)
```

```python
contents1=soup.find_all('table',class_='infobox vevent')
for i in contents1:
    print(i.text)
```

```python
contents2=soup.find_all("div",class_='plainlist')
for i in contents2:
    print(i.text)
```
